```python
import xarray as xr
import matplotlib.pyplot as plt
import data_catalog
import tseries_mod
import esmlab
```

```python
data_catalog.set_catalog('experiments')
```

```python
experiment = 'piControl-cmip5'
piControl_cmip5 = xr.Dataset()
for varname in ['SFCO2_LND', 'SFCO2_OCN']:
    da = tseries_mod.tseries_get(varname, 'atm', experiment, stream='cam2.h0')
    piControl_cmip5 = xr.merge((piControl_cmip5, da))
piControl_cmip5_ann = esmlab.climatology.compute_ann_mean(piControl_cmip5)
piControl_cmip5_ann.time.attrs['calendar'] = piControl_cmip5.time.attrs['calendar']
for varname in ['SFCO2_LND', 'SFCO2_OCN']:
    da = piControl_cmip5_ann[varname].cumsum(dim='time', keep_attrs=True)
    da.name = varname+'_CUM'
    da.attrs['units'] = 'Pg'
    piControl_cmip5_ann = xr.merge((piControl_cmip5_ann, da))
print(piControl_cmip5_ann)

experiment = 'esm-piControl-cmip5'
esm_piControl_cmip5 = xr.Dataset()
for varname in ['SFCO2_LND', 'SFCO2_OCN']:
    da = tseries_mod.tseries_get(varname, 'atm', experiment, stream='cam2.h0')
    esm_piControl_cmip5 = xr.merge((esm_piControl_cmip5, da))
esm_piControl_cmip5_ann = esmlab.climatology.compute_ann_mean(esm_piControl_cmip5)
esm_piControl_cmip5_ann.time.attrs['calendar'] = esm_piControl_cmip5.time.attrs['calendar']
for varname in ['SFCO2_LND', 'SFCO2_OCN']:
    da = esm_piControl_cmip5_ann[varname].cumsum(dim='time', keep_attrs=True)
    da.name = varname+'_CUM'
    da.attrs['units'] = 'Pg'
    esm_piControl_cmip5_ann = xr.merge((esm_piControl_cmip5_ann, da))
print(esm_piControl_cmip5_ann)

experiment = 'piControl'
piControl = xr.Dataset()
for varname in ['SFCO2_LND', 'SFCO2_OCN']:
    da = tseries_mod.tseries_get(varname, 'atm', experiment)
    piControl = xr.merge((piControl, da))
piControl_ann = esmlab.climatology.compute_ann_mean(piControl)
piControl_ann.time.attrs['calendar'] = piControl.time.attrs['calendar']
for varname in ['SFCO2_LND', 'SFCO2_OCN']:
    da = piControl_ann[varname].cumsum(dim='time', keep_attrs=True)
    da.name = varname+'_CUM'
    da.attrs['units'] = 'Pg'
    piControl_ann = xr.merge((piControl_ann, da))
print(piControl_ann)
```

```python
varname = 'SFCO2_LND_CUM'
tseries_mod.tseries_plot_1var(varname, [piControl_cmip5_ann, esm_piControl_cmip5_ann, piControl_ann],
                              ['piControl CESM1', 'esm-piControl CESM1', 'piControl CESM2'],
                              varname, region_val='Global')
```

```python
varname = 'SFCO2_OCN_CUM'
tseries_mod.tseries_plot_1var(varname, [piControl_cmip5_ann, esm_piControl_cmip5_ann, piControl_ann],
                              ['piControl CESM1', 'esm-piControl CESM1', 'piControl CESM2'],
                              varname, region_val='Global')
```
