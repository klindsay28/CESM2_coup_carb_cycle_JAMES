```python
import xarray as xr
import matplotlib.pyplot as plt
import data_catalog
import tseries_mod
import esmlab_wrap
```

```python
data_catalog.set_catalog('experiments')
```

```python
varnames = ['SFCO2_LND', 'SFCO2_OCN']
```

```python
experiment = 'piControl-cmip5'
piControl_cmip5 = tseries_mod.tseries_get_vars(varnames, 'atm', experiment, stream='cam2.h0')
piControl_cmip5_ann = esmlab_wrap.compute_ann_mean(piControl_cmip5)
print(piControl_cmip5_ann)
```

```python
experiment = 'esm-piControl-cmip5'
esm_piControl_cmip5 = tseries_mod.tseries_get_vars(varnames, 'atm', experiment, stream='cam2.h0')
esm_piControl_cmip5_ann = esmlab_wrap.compute_ann_mean(esm_piControl_cmip5)
print(esm_piControl_cmip5_ann)
```

```python
experiment = 'piControl'
piControl = tseries_mod.tseries_get_vars(varnames, 'atm', experiment)
piControl_ann = esmlab_wrap.compute_ann_mean(piControl)
print(piControl_ann)
```

```python
experiment = 'esm-piControl'
esm_piControl = tseries_mod.tseries_get_vars(varnames, 'atm', experiment)
esm_piControl_ann = esmlab_wrap.compute_ann_mean(esm_piControl)
print(esm_piControl_ann)
```

```python
experiment = 'piControl-2deg'
piControl_2deg = tseries_mod.tseries_get_vars(varnames, 'atm', experiment)
piControl_2deg_ann = esmlab_wrap.compute_ann_mean(piControl_2deg)
print(piControl_2deg_ann)
```

```python
# add cummulate flux vars to datasets
for ds in [piControl_cmip5_ann, esm_piControl_cmip5_ann, piControl_ann, esm_piControl_ann, piControl_2deg_ann]:
    for varname in ['SFCO2_LND', 'SFCO2_OCN']:
        da = ds[varname].cumsum(dim='time', keep_attrs=True)
        da.name = varname+'_CUM'
        da.attrs['units'] = 'Pg'
        ds[da.name] = da
```

```python
ds_list = [piControl_ann, esm_piControl_ann, piControl_cmip5_ann, esm_piControl_cmip5_ann]
legend_list = ['piControl CESM2', 'esm-piControl CESM2', 'piControl CESM1', 'esm-piControl CESM1']
varname = 'SFCO2_LND_CUM'
tseries_mod.tseries_plot_1var(varname, ds_list, legend_list, varname, region_val='Global')
varname = 'SFCO2_OCN_CUM'
tseries_mod.tseries_plot_1var(varname, ds_list, legend_list, varname, region_val='Global')
```

```python
ds_list = [piControl_ann, esm_piControl_ann, piControl_2deg_ann]
legend_list = ['piControl CESM2', 'esm-piControl CESM2', 'piControl CESM2 2deg']
varname = 'SFCO2_LND_CUM'
tseries_mod.tseries_plot_1var(varname, ds_list, legend_list, varname, region_val='Global')
varname = 'SFCO2_OCN_CUM'
tseries_mod.tseries_plot_1var(varname, ds_list, legend_list, varname, region_val='Global')
```
